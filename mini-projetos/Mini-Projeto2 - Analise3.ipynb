```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

sns.set_style('darkgrid')
plt.figure(figsize = (14, 10))
fig = sns.barplot(df.fuelType, df.price, df.gearbox, palette = 'colorblind')
fig.set_title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio', fontsize = 18, pad = 20)
fig.set_xlabel('Tipo de combustível', fontsize = 18, labelpad = 20)
fig.set_ylabel('Preço médio', fontsize = 18, labelpad = 20)
```

```python
# Salvando o plot
fig.figure.savefig("plots/Analise3/fueltype-vehicleType-price-exercicio.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

sns.set_style('darkgrid')
plt.figure(figsize = (14, 10))
fig = sns.barplot(df.vehicleType, df.powerPS, df.gearbox, palette = 'bright')
fig.set_title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio', fontsize = 18, pad = 20)
fig.set_xlabel('Tipo de veículo', fontsize = 18, labelpad = 20)
fig.set_ylabel('Potência média', fontsize = 18, labelpad = 20)
```

```python
# Salvando o plot
fig.figure.savefig("plots/Analise3/vehicletype-fueltype-power-exercicio.png")
```
