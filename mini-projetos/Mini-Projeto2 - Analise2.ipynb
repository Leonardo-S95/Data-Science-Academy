```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Número de veículos pertencentes a cada marca

```python
# Crie um Plot que mostre o número de veículos pertencentes a cada marca

sns.set_style('whitegrid')
g = sns.catplot(y = 'brand', data = df, kind = 'count', height = 12, palette = 'viridis')
g.ax.set_title('Número de veículos por marca', fontsize = 18, pad = 20)
g.ax.set_xlabel('Quantidade de veículos', fontsize = 18, labelpad = 20)
g.ax.set_ylabel('Marca', fontsize = 18, labelpad = 20)
```

```python
# Salvando o plot
g.savefig(("plots/Analise2/brand-vehicleCount-exercicio.png"))
```

## Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

```python
# Crie um Plot com o Preço médio dos veículos com base no tipo de veículo, bem como no tipo de caixa de câmbio

plt.figure(figsize = [14, 10])
fig = sns.barplot(df.vehicleType, df.price, df.gearbox, palette = 'Blues')
fig.set_title('Preço médio dos veículos com base no tipo de veículo e tipo de caixa de câmbio', fontsize = 18, pad = 20)
fig.set_xlabel('Tipo de veículo', fontsize = 18, labelpad = 20)
fig.set_ylabel('Preço médio', fontsize = 18, labelpad = 20)
```

```python
# Salvando o plot
fig.figure.savefig("plots/Analise2/vehicletype-gearbox-price-exercicio.png")
```
