## Análise 1

```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style = "white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding = "latin-1")
```

## Distribuição de Veículos com base no Ano de Registro

```python
# Crie um Plot com a Distribuição de Veículos com base no Ano de Registro

fig = plt.figure(figsize = (12, 10))
sns.distplot(df['yearOfRegistration'])
plt.xlabel('Year of registration', fontsize = 18, labelpad = 20)
plt.ylabel('Density', fontsize = 18, labelpad = 20)
plt.title('Vehicle distribution based on registration year', fontsize = 18, pad = 20)
plt.show()
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/vehicle-distribution-exercicio.png")
```

## Variação da faixa de preço pelo tipo de veículo

```python
# Crie um Boxplot para avaliar os outliers

fig = plt.figure(figsize = (12, 10))
boxplot = sns.boxplot(df['vehicleType'], df['price'])
boxplot.set_xlabel('Tipo de veículo', fontsize = 18, labelpad = 20)
boxplot.set_ylabel('Preço', fontsize = 18, labelpad = 20)
plt.title('Variação de faixa de preço pelo tipo de veículo', fontsize = 18, pad = 20)
```

```python
# Salvando o plot
fig.savefig("plots/Analise1/price-vehicleType-boxplot-exercicio.png")
```

## Contagem total de veículos à venda conforme o tipo de veículo

```python
# Crie um Count Plot que mostre o número de veículos pertencentes a cada categoria 

g = plt.figure(figsize = [12, 10])
sns.countplot(df['vehicleType'])
plt.title('Número de veículos pertencentes a cada categoria', fontsize = 18, pad = 20)
plt.xlabel('Tipo de veículo', fontsize = 18, labelpad = 20)
plt.ylabel('Quantidade', fontsize = 18, labelpad = 20)
```

```python
# Salvando o plot
g.savefig("plots/Analise1/count-vehicleType-exercicio.png")
```
